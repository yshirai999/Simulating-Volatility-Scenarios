# Experiments using Transformer Architecture

```python
import torch
import torch.nn as nn
import numpy as np
from Linear import Linear2D, Linear
from sklearn.preprocessing import StandardScaler
# Base Env params

# need to make feature and target steps the same for the embedding
feature_steps =  1008
target_steps = 1
quantized = False
num_epochs = 100
n_clusters = 500
quantiles=(0.25,0.75)
device = "cpu" #'cuda' # 'cuda' or 'cpu'. For 'cuda', requires GPU compatible with CUDA version 11.8 or higher, and PyTorch version 2.0 or higher.
```

```python
epoch_list = [10, 25, 50, 100]
input_dim = 2
output_dim = 4
for e in epoch_list:
    num_epochs = e
    model = Linear2D(tickers=['spy'], target_steps=target_steps, feature_steps=feature_steps, scaler=StandardScaler,
        input_dim=input_dim, output_dim=output_dim, quantized=quantized,
        n_clusters=n_clusters, num_epochs=num_epochs, quantiles=quantiles, device=device)
    print(f"Training for {e} epochs")
    model.train()
    model.predict()
    model.plot_quantile_predicitons(t="spy", title=f"num_epochs={num_epochs} Quantized:{quantized}")
    print("-"*50)
```
